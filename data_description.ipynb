```python
btc_data.rename(columns={'Open time': 'OpenTime', 'Close time': 'CloseTime', 'Quote asset volume': 'QuoteAssetVolume', 'Number of trades':'NumberOfTrades', 'Taker buy base asset volume':'TakerBuyBaseAssetVolume','Taker buy quote asset volume': 'TakerBuyQuoteAssetVolume'  }, inplace=True)
btc_data.tail()
```

```python
# btc_data['OpenTime'].describe()
# btc_data['OpenTime'] = pd.to_datetime(btc_data['OpenTime'])
btc_data['date'] = pd.to_datetime(btc_data['OpenTime'],unit='ms')
# btc_data.astype({'OpenTime': 'datetime64[ms]'})
```

```python
btc_data['date'].dt.hour.unique()
(btc_data['date'].dt.hour==0).unique()
```

```python
btc_data_monday = btc_data[(btc_data['date'].dt.day_name()=='Monday')] 
btc_moday_start = btc_data_monday[btc_data_monday['date'].dt.hour==0]
btc_moday_start.count()
```

```python
btc_moday_start['WeeklyClose']=btc_moday_start['Open'].shift(-1)
btc_moday_start['WeeklyChange']=btc_moday_start['WeeklyClose']-btc_moday_start['Open']
btc_moday_start
```

```python
btc_moday_start["WeeklyChange"].plot()
```
